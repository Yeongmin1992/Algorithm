```python
# 일반적인 형태 #1
def function(입력):
    if 입력 > 일정값: # 입력이 일정 값 이상이면
        return function(입력 - 1) # 입력보다 작은 값
    else:
        return 일정값, 입력값, 또는 특정값 # 재귀 호출 종료
```

```python
def factorial(num):
    if num > 1:
        return num * factorial(num - 1)
    else:
        return num
```

```python
for num in range(10):
    print(factorial(num))
```

#### 재귀는 스택구조

#### 1부터 num까지의 곱 출력하기

```python
# 비재귀
def multiple(num):
    return_value = 1
    for i in range(1, num + 1):
        return_value = return_value * i
    return return_value
```

```python
# 재귀
def multiple(num):
    if num <= 1:
        return num
    return num * multiple(num - 1)
```

```python
print(multiple(10))
```

#### 숫자가 들어있는 리스트가 주어졌을 때, 리스트의 합을 리턴하는 함수

```python
def sum_list(data):
    if len(data) <= 1:
        return data[0]
    return data[0] + sum_list(data[1:])
```

```python
import random
data = random.sample(range(100), 10)
data
```

```python
sum_list(data)
```

#### 거꾸로 읽어도 동일한 단어(회문) 판별

```python
def same(word):
    if len(word) <= 1:
        return True
    if word[0] == word[-1]:
        return same(word[1:-1])
    else:
        return False
```

```python
same("level")
```

#### 정수 n에 대해 n이 1이 될 때까지 n이 홀수이면 3 * n + 1, 짝수이면 n / 2를 반복하고 각 과정의 결과값 프린트

```python
def func(n):
    if n == 1:
        print(1)
        return
    if n % 2 == 1:
        print(n)
        n = 3 * n + 1
        func(n)
    else:
        print(n)
        n = n / 2
        func(n)
```

```python
func(3)
```

```python
def func(n):
    print(n)
    if n == 1:
        return n
    if n % 2 == 1:
        return(func((3 * n) + 1))
    else:
        return(func(int(n / 2)))
```

```python
func(3)
```

```python
def func(data):
    if data == 1:
        return 1
    elif data == 2:
        return 2
    elif data == 3:
        return 4
    
    return func(data - 1) + func(data - 2) + func(data - 3)
```
